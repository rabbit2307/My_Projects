```python
def classify_object(model, ra, dec, u, g, r, i, z, run, rerun, camcol, field, redshift, plate, mjd, fiberid):
    
    input_data = np.array([[ra, dec, u, g, r, i, z, run, rerun, camcol, field, redshift, plate, mjd, fiberid]])
    input_data_scaled = scaler.transform(input_data)

    prediction = model.predict(input_data_scaled)

    classes = {0: "Galaxy", 1: "Star", 2: "Quasar"}

    return classes[prediction[0]]

sample_ra = 183.6802074
sample_dec = 0.126185092
sample_u = 19.38298
sample_g = 18.19169
sample_r = 17.47428
sample_i = 17.08732
sample_z = 16.80125
sample_run = 752
sample_rerun = 301
sample_camcol = 4
sample_field = 268
sample_redshift = 0.1231112
sample_plate = 287
sample_mjd = 52023
sample_fiberid = 513

predicted_class_for_sample = classify_object(model,
                                             ra=sample_ra,
                                             dec=sample_dec,
                                             u=sample_u,
                                             g=sample_g,
                                             r=sample_r,
                                             i=sample_i,
                                             z=sample_z,
                                             run=sample_run,
                                             rerun=sample_rerun,
                                             camcol=sample_camcol,
                                             field=sample_field,
                                             redshift=sample_redshift,
                                             plate=sample_plate,
                                             mjd=sample_mjd,
                                             fiberid=sample_fiberid
                                            )

print(f"The predicted class for the sample object is: {predicted_class_for_sample}")
```

## PLC Visulisation

PCA: Class Separation plot visualizes the dataset by projecting it onto its first two principal components (PC1 and PC2). Principal Component Analysis (PCA) is employed as a dimensionality reduction technique to transform high-dimensional data into a lower-dimensional representation while preserving as much of the original variance as possible. Its primary purpose here is to enable visualization of complex datasets in two dimensions, allowing us to visually inspect how well different classes are separated and identify potential linear separability for classification tasks, as well as to simplify data for models that benefit from reduced feature sets.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(10,7))
for label in np.unique(y):
    plt.scatter(X_pca[y == label, 0], X_pca[y == label, 1], label=le.inverse_transform([label])[0], alpha=0.5)
plt.title("PCA: Class Separation")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.show()
```

The distribution of the galaxy, quaser, and star classes within the reduced PC1-PC2 space. There appears to be significant overlap between the 'Galaxy' and 'Star' clusters, indicating that these two classes are not linearly separable based solely on their first two principal components. While the 'QSO' class shows a slightly more distinct clustering towards the right side of the plot, it still exhibits some degree of intermixing with both 'Galaxy' and 'Star' points. This visual assessment suggests that projecting the data onto just two principal components does not achieve clear linear separation for all classes, implying that a simple linear classifier might face challenges in perfectly distinguishing these astronomical objects if relying solely on these two principal components.

## t-SNE Visualization

This t-SNE Class Separation plot visualizes the dataset using t Distributed Stochastic Neighbor Embedding t-SNE, a non linear dimensionality reduction technique. Unlike PCA, t-SNE focuses on preserving the local structure of the data, aiming to keep similar data points close together and dissimilar points far apart in the lower dimensional space. This makes t-SNE particularly effective for visualizing high dimensional data and revealing natural clusters or complex non linear class separations not apparent with linear methods.

```python


from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

plt.figure(figsize=(10,7))
for label in np.unique(y):
    plt.scatter(X_tsne[y == label, 0], X_tsne[y == label, 1], label=le.inverse_transform([label])[0], alpha=0.5)
plt.title("t-SNE: Class Separation")
plt.xlabel("Dim 1")
plt.ylabel("Dim 2")
plt.legend()
plt.show()
```

t-SNE provides much clearer visual separation for the three classes Galaxy, QSO, and Star compared to PCA. The Star class forms a prominent and largely distinct cluster on the left and upper left. The QSO class appears as several well defined, though smaller, clusters towards the center and right, showing better isolation from the other two. Despite some intermixing, especially between Galaxy and Star points in central regions, the overall clustering is more pronounced, suggesting these astronomical object classes are more separable through non linear boundaries.

## Training Advanced Models

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=128)
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=128)
svm_classifier = SVC(kernel='rbf', C=1.0, gamma='scale', probability=True, random_state=128)
mlp_classifier = MLPClassifier(hidden_layer_sizes=(100, 50), activation='relu', solver='adam', max_iter=500, random_state=128)

rf_model.fit(X_train, y_train)
xgb_model.fit(X_train, y_train)
svm_classifier.fit(X_train, y_train)
mlp_classifier.fit(X_train, y_train)
```

```python
models = {
    'RandomForest': RandomForestClassifier(),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'),
    'SVM': SVC(),
    'MLP': MLPClassifier()
}
```

## Hyperparameter Tuning

```python
param_grids = {
    'RandomForest': {
        'n_estimators': [100, 300, 500],
        'max_depth': [10, 20, 40, None],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'bootstrap': [True, False]
    },
    'XGBoost': {
        'n_estimators': [100, 300, 500],
        'max_depth': [3, 6, 10],
        'learning_rate': [0.01, 0.05, 0.1],
        'subsample': [0.7, 0.9, 1],
        'colsample_bytree': [0.7, 0.9, 1],
        'gamma': [0, 0.1, 0.3]
    },
    'SVM': {
        'C': [0.1, 1, 10, 100],
        'gamma': ['scale', 'auto', 0.001, 0.01, 0.1],
        'kernel': ['rbf', 'poly', 'linear']
    },
    'MLP': {
        'hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 50)],
        'activation': ['relu', 'tanh'],
        'solver': ['adam', 'sgd'],
        'alpha': [0.0001, 0.001, 0.01],
        'learning_rate': ['constant', 'adaptive'],
        'max_iter': [300]
    }
}
```

```python
best_models = {}

for name in models:
    print(f"\n Tuning {name}...")
    
    grid = GridSearchCV(
        estimator=models[name],
        param_grid=param_grids[name],
        cv=3,
        n_jobs=-1,
        verbose=2
    )
    
    grid.fit(X_train, y_train)
    print(f"Best parameters for {name}: {grid.best_params_}")
    preds = grid.predict(X_test)
    print("\nClassification Report:")
    print(classification_report(y_test, preds))
    
    print("\nConfusion Matrix:")
    print(confusion_matrix(y_test, preds))
    
    best_models[name] = grid.best_estimator_
```

The hyperparameter tuning revealed that Random Forest, XGBoost,Neural Network (MLP) and SVM models all achieved outstanding 0.99 accuracy. A deeper look at the classification reports and confusion matrices indicates that Random Forest and XGBoost provided the most balanced and consistently near-perfect precision and recall across all classes, marginally outperforming SVM and MLP, despite all four models demonstrating highly effective classification capabilities after optimization.

## EVALUATION FUNCTION

```python
def preprocess_new_data(file_path, le, scaler, feature_names):
    if not os.path.exists(file_path):
        print(f"File not found at: {file_path}")
        return None, None, None

    data = pd.read_csv(file_path)
    data.drop(columns=[col for col in ['objid', 'specobjid'] if col in data.columns], inplace=True) 

    if 'class' in data.columns: 
        unknown_classes = set(data['class']) - set(le.classes_)
        if unknown_classes:
            data = data[~data['class'].isin(unknown_classes)] 

        data['class_encoded'] = le.transform(data['class']) 
        y = data['class_encoded'] 
        X = data.drop(columns=['class', 'class_encoded'])
    else:
        y = None
        X = data 

    try:
        X = X[feature_names]
    except KeyError as e:
        print(f"Feature mismatch: {e}")
        return None, None, None

    X_scaled = scaler.transform(X)
    return X_scaled, y, data


def evaluate_model_on_new_data(name, model, X_scaled, y_true, le):
    preds = model.predict(X_scaled)
    class_counts = collections.Counter(preds)
    readable_counts = {le.inverse_transform([k])[0]: v for k, v in class_counts.items()}
    print(f"\n{name} - Prediction Distribution:", readable_counts)

    if y_true is not None:
        acc = accuracy_score(y_true, preds)
        print(f"{name} Accuracy: {acc:.4f}")
        print(classification_report(y_true, preds, target_names=le.classes_))

        cm = confusion_matrix(y_true, preds)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=le.classes_, yticklabels=le.classes_)
        plt.title(f"Confusion Matrix - {name}")
        plt.xlabel("Predicted")
        plt.ylabel("Actual")
        plt.tight_layout()
        plt.show()

def evaluate_all_models(file_path, models_dict, le, scaler, feature_names):
    X_scaled, y_true, _ = preprocess_new_data(file_path, le, scaler, feature_names)
    if X_scaled is None:
        return
    for name, model in models_dict.items():
        evaluate_model_on_new_data(name, model, X_scaled, y_true, le)
```

## EVALUATION ON NEW DATA

```python
new_data_file_path = r"C:\Users\Lenovo\Documents\My_Projects\ML Projects\Skyserver_12_30_2019 4_49_58 PM.csv"

models = {
    "Random Forest": rf_model,
    "XGBoost": xgb_model,
    "SVM": svm_classifier,
    "Neural Network (MLP)": mlp_classifier
}

evaluate_all_models(
    file_path=new_data_file_path,
    models_dict=models,
    le=le,
    scaler=scaler,
    feature_names=feature_names
)
```

This stark difference highlights that Random Forest and XGBoost generalized far more effectively to unseen data compared to SVM and MLP.
